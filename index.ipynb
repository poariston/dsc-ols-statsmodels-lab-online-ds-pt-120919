## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
import statsmodels.api as sm  
import statsmodels.formula.api as smf

# build the formula 
formula = 'sales ~ TV'
# create a fitted model in one line
model = smf.ols(formula=formula,data = df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

```python
# comments on "Goodness of fit" here

# The R-squared value is 0.61 i.e. 61% of the variance in the target variable can be explained by TV spending. 
# The slope: A "unit" increase in TV spending is associated with a 0.0475 "unit" 
# increase in Sales or an additional $1,000 spent on TV is associated with an increase in sales of $47.5
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
X_new=pd.DataFrame({'TV':[df.TV.min(),df.TV.max()]})
print(X_new)

# make predictions for those x values and store them
preds=model.predict(X_new)
print('\n',preds)

# X_new['sales'] = preds
# print('\n',X_new)

# first, plot the observed data and the least squares line


sns.lmplot(x='TV', y ='sales',data=df,fit_reg=False)
plt.plot(X_new, preds, c='red', linewidth=2)
plt.show();
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig= plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model,'TV',fig = fig)
plt.show();
```

```python
# Record Your observations on heteroscedasticity
# From the first and second plot in the first row, we see a cone-shape which is a sign of heteroscedasticity. 
# i.e. the residuals are heteroscedastic. This violates an assumption.
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals=model.resid
sm.graphics.qqplot(residuals,dist=stats.norm,line='45',fit=True)
fig.show()
```

```python
# Record Your observations on the normality assumption
# With a pretty good QQ plot the normality assumption of the residuals seems fulfilled.
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
# model
# build the formula 
formula2 = 'sales ~ radio'
# create a fitted model in one line
model2 = smf.ols(formula=formula2,data = df).fit()
```

```python
# Get Regression Diagnostics Summary
model2.summary()
```

```python
# Draw a prediction line with data points on a scatter plot for X and Y
# create a DataFrame with the minimum and maximum values of x
X_new=pd.DataFrame({'radio':[df.radio.min(),df.radio.max()]})
print(X_new)

# make predictions for those x values and store them
preds2=model2.predict(X_new)
print('\n',preds2)

# first, plot the observed data, then the least squares line
sns.lmplot(x='radio', y ='sales',data=df,fit_reg=False,height=5,aspect=1.62)
plt.plot(X_new, preds2, c='red', linewidth=2,label = 'best fit regression line from OLS')
plt.scatter(x=df.radio.mean(),y=df.sales.mean(),linewidth=5,label = '(x mean,y mean) point')
plt.legend()
plt.show();
```

```python
# Visualize the error term for variance and homoscedasticity
fig= plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model2,'radio',fig = fig)
plt.show();
```

```python
# Check the normality assumptions by creating a QQ-plot
residuals2=model2.resid
sm.graphics.qqplot(residuals2,dist=stats.norm,line='45',fit=True)
fig.show()
```

```python
# Record your observations here for goodnes of fit 
# comments on "Goodness of fit" 
# The R-squared value is 0.332 i.e. 33% of the variance in the target variable can be explained by radio spending. 
# The slope: A "unit" increase in radio spending is associated with a 0.0475 "unit" 
# increase in Sales or an additional $1,000 spent on TV is associated with an increase in sales of $20

# observations on heteroscedasticity
# From the first and second plot in the first row, we see a cone-shape which is a sign of heteroscedasticity. 
# i.e. the residuals are heteroscedastic. This violates an assumption.

#  observations on the normality assumption
# With deviations from the ideal diagonal line in QQ plot, the normality assumption of the residuals is not fulfilled.
# compared to a normal distribution, the residuals' distribution is skewed. 
# For the low values, when the normal distrib reaches 10 % fo the values, the res distrib is still at a lower quantile.
# As forthe high values, when we are close to 100% of the points 
# for the normal distribution, we are still below 100% for the residuals distribution. 
# so there are more data towards the high value for the res distrib. 
```
